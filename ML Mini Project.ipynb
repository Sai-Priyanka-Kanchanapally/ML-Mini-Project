# Splitting the Data Sets Into X and Y by the dependent and independent variables (data selected by PCA

```python

X=pca_df_train
y=Scaled_df_train['Consumer disputed?']
```

```python
# Importing libraries
!pip install xgboost
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score
```

```python
# splitting training and test dataset
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.3,random_state=0,stratify=y)
```

```python
# Model deployment
models= {'LogisticRegression': LogisticRegression(),'DecisionTreeClassifier': DecisionTreeClassifier(),'RandomForestClassifier':RandomForestClassifier(),'AdaBoostClassifier': AdaBoostClassifier(),'GradientBoostingClassifier':GradientBoostingClassifier(),'KNeighborsClassifier':KNeighborsClassifier(),'XGBClassifier':XGBClassifier()}
```

```python
models = {'LogisticRegression': LogisticRegression(),
          'DecisionTreeClassifier': DecisionTreeClassifier(),
          'RandomForestClassifier': RandomForestClassifier(),
          'AdaBoostClassifier': AdaBoostClassifier(),
          'GradientBoostingClassifier': GradientBoostingClassifier(),
          'KNeighborsClassifier': KNeighborsClassifier(),
          'XGBClassifier': XGBClassifier()}
```

```python
y=y.replace({'No': 0, 'Yes': 1})
y_train = y_train.replace({'No': 0, 'Yes': 1})
y_test = y_test.replace({'No': 0, 'Yes': 1})
```

```python
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy of {name} is {round(accuracy*100, 2)}%")
```

```python
for name, model in models.items():
    scores = cross_val_score(model, X_train, y_train, cv=5)
    print(f"Accuracy of {name} is {round(scores.mean()*100, 2)}%")
```
